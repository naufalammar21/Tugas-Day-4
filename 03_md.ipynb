Ganjil atau Genap?

Buatlah script untuk menentukan suatu angka merupakan bilangan ganjil atau genap.

Contoh:

Input: 43

Output: Ganjil

Input: 1032

Output: Genap


```python
print('=======================Program Menentukan Bilangan Ganjil Genap======================\n')
bil = int(input('masukkan bilangan : '))

if bil%2==0:
  print('Bilangan adalah bilangan genap')
else:
  print('Bilangan tersebut adalah ganjil')
```

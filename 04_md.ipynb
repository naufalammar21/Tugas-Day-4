Kabisat

Buatlah script untuk menentukan suatu tahun merupakan tahun kabisat atau bukan.

Contoh:

Input: 1900

Output: Bukan kabisat

Input: 2000

Output: Kabisat

Input: 2001

Output: Bukan kabisat

Input: 2016

Output: Kabisat


```python
print('=======================Program Menentukan Tahun Kabisat======================\n')
tahun = int(input('input tahun : '))

if tahun%4==0:
  print('tahun tersebut adalah tahun kabisat')
else:
  print('tahun tersebut bukan kabisat')
```
